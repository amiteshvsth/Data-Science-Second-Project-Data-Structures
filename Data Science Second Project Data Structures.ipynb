```python
# Q1. Discuss String Slicing and provide examples.

# Answer

'''

String slicing in Python allows you to access a substring of a given string. The syntax for slicing is:


string[start:stop:step]
Here's what each parameter means:

start: The index where the slice begins (inclusive). Defaults to 0.
stop: The index where the slice ends (exclusive). If not provided, defaults to the length of the string.
step: The step size or stride, which determines the increment between each index. Defaults to 1.


examples:

Basic Slicing:
text = "Hello, World!"
print(text[0:5])  # Output: Hello


Slicing with Default Values:
print(text[:5])   # Output: Hello (same as text[0:5])
print(text[7:])   # Output: World! (starts from index 7 to the end)


Negative Indices:
Copy code
print(text[-6:])  # Output: World! (starts 6 positions from the end)
print(text[:-7])  # Output: Hello, (up to but not including the last 7 characters)


Using Step:
print(text[::2])  # Output: Hlo ol! (every second character)
print(text[1::2]) # Output: el,Wrd (every second character starting from index 1)


Reversing a String:
print(text[::-1]) # Output: !dlroW ,olleH (reverses the string)


Combining Slicing Parameters:
print(text[1:9:2]) # Output: el,W (from index 1 to 8, with step 2)

'''
```

```python
# Q3. Describe how to access, modify and delete elements in a list with examples.

# Answer

'''

1. Accessing Elements
You can access elements in a list using indexing and slicing.

Indexing:
my_list = [10, 20, 30, 40, 50]
print(my_list[0])  # Output: 10
print(my_list[2])  # Output: 30
print(my_list[-1]) # Output: 50 (last element)


Slicing:
print(my_list[1:4])   # Output: [20, 30, 40]
print(my_list[:3])    # Output: [10, 20, 30]
print(my_list[2:])    # Output: [30, 40, 50]
print(my_list[::2])   # Output: [10, 30, 50] (every second element)



2. Modifying Elements
You can modify elements in a list by assigning a new value to a specific index or range of indices.

Modifying a Single Element:
my_list[1] = 25
print(my_list)  # Output: [10, 25, 30, 40, 50]


Modifying Multiple Elements:
my_list[1:3] = [26, 36]
print(my_list)  # Output: [10, 26, 36, 40, 50]


Using List Methods:
append(): Adds an element to the end.
my_list.append(60)
print(my_list)  # Output: [10, 26, 36, 40, 50, 60]

extend(): Adds multiple elements to the end.
my_list.extend([70, 80])
print(my_list)  # Output: [10, 26, 36, 40, 50, 60, 70, 80]

insert(): Inserts an element at a specified position.
my_list.insert(2, 35)
print(my_list)  # Output: [10, 26, 35, 36, 40, 50, 60, 70, 80]


3. Deleting Elements
You can delete elements using the del statement, remove(), pop(), and clear() methods.

Using del Statement:
del my_list[2]
print(my_list)  # Output: [10, 26, 36, 40, 50, 60, 70, 80]

del my_list[1:3]
print(my_list)  # Output: [10, 40, 50, 60, 70, 80]


Using remove() Method:
my_list.remove(40)
print(my_list)  # Output: [10, 50, 60, 70, 80]

Using pop() Method:
my_list.pop()    # Removes and returns the last element.
print(my_list)   # Output: [10, 50, 60, 70]

my_list.pop(1)   # Removes and returns the element at index 1.
print(my_list)   # Output: [10, 60, 70]


Using clear() Method:
my_list.clear()
print(my_list)   # Output: []

'''
```

```python
# Q4. Compare and contrast tuples and lists with examples.

# Answer

'''

Tuples and lists are both sequence data types in Python that can store a collection of items. However, they have several key differences that affect how and when you might use each.
Hereâ€™s a comparison:

1. Mutability
Lists: Lists are mutable, meaning you can change their content after creation by adding, removing, or modifying elements.
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)  # Output: [10, 2, 3]

Tuples: Tuples are immutable, meaning once they are created, their content cannot be changed.
my_tuple = (1, 2, 3)
# my_tuple[0] = 10  # This will raise a TypeError


2. Syntax
Lists: Lists are defined using square brackets [].
my_list = [1, 2, 3]

Tuples: Tuples are defined using parentheses ().
my_tuple = (1, 2, 3)


3. Performance
Lists: Lists are generally slower than tuples when it comes to iteration due to their mutability.
Tuples: Tuples can be more efficient in terms of performance, especially when dealing with large datasets that do not require modification.


4.Use Cases
Lists: Use lists when you need a collection of items that can change over time.
shopping_list = ["apple", "banana", "cherry"]
shopping_list.append("date")
print(shopping_list)  # Output: ["apple", "banana", "cherry", "date"]

Tuples: Use tuples when you need a collection of items that should not change. Tuples are often used for fixed collections of items, such as coordinates or settings.
coordinates = (10.0, 20.0)
print(coordinates)  # Output: (10.0, 20.0)


5. Methods
Lists: Lists have numerous built-in methods for modifying their content (append(), extend(), insert(), remove(), pop(), clear(), sort(), reverse(), etc.).
my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

Tuples: Tuples have fewer methods since they are immutable. The main methods are count() and index().
my_tuple = (1, 2, 3, 2)
print(my_tuple.count(2))  # Output: 2
print(my_tuple.index(3))  # Output: 2


Nesting and Heterogeneity
Lists and Tuples: Both lists and tuples can contain mixed data types and other lists or tuples.
mixed_list = [1, "apple", (2, 3)]
mixed_tuple = (1, "apple", [2, 3])
print(mixed_list)   # Output: [1, "apple", (2, 3)]
print(mixed_tuple)  # Output: (1, "apple", [2, 3])

Examples
List Example:
my_list = [1, 2, 3]
my_list.append(4)
my_list[0] = 10
print(my_list)  # Output: [10, 2, 3, 4]


Tuple Example:
my_tuple = (1, 2, 3)
print(my_tuple[0])  # Output: 1
# my_tuple[0] = 10  # Raises TypeError: 'tuple' object does not support item assignment


Using Lists for Mutable Collections:
fruits = ["apple", "banana", "cherry"]
fruits.append("date")
print(fruits)  # Output: ["apple", "banana", "cherry", "date"]


Using Tuples for Immutable Collections:
point = (3.5, 7.2)
print(point)  # Output: (3.5, 7.2)


Summary
Mutability: Lists are mutable; tuples are immutable.
Syntax: Lists use []; tuples use ().
Performance: Tuples are generally faster and more memory-efficient than lists.
Use Cases: Lists are used for collections that change; tuples are used for fixed collections.
Methods: Lists have more methods due to mutability; tuples have fewer methods.
Choosing between a list and a tuple depends on whether you need to modify the collection of items or not. If you need mutability, use a list. 
If you need immutability, use a tuple.


'''
```

```python
# Q5. Describe the key features of sets and provide examples of their use.

# Answer

'''

Sets are a built-in data structure in Python that provide a way to store collections of unique elements. They are particularly useful for operations involving membership tests, 
removing duplicates, and performing mathematical set operations like union, intersection, and difference. Here are the key features of sets:

Key Features of Sets
Unordered Collection: Sets do not maintain any order. The elements might appear in a different order every time you access the set.
Unique Elements: Sets automatically handle duplicate elements. Any duplicates are removed.
Mutable: Sets can be modified by adding or removing elements, but the elements themselves must be immutable (e.g., numbers, strings, tuples).
No Indexing or Slicing: Sets do not support indexing or slicing because they are unordered.
Set Operations: Sets support mathematical operations like union, intersection, difference, and symmetric difference.


Examples of Set Operations
Creating a Set:
my_set = {1, 2, 3, 4, 5}
print(my_set)  # Output: {1, 2, 3, 4, 5}
empty_set = set()  # Creates an empty set


Adding Elements:
my_set.add(6)
print(my_set)  # Output: {1, 2, 3, 4, 5, 6}
my_set.update([7, 8, 9])
print(my_set)  # Output: {1, 2, 3, 4, 5, 6, 7, 8, 9}


Removing Elements:
my_set.remove(3)
print(my_set)  # Output: {1, 2, 4, 5, 6, 7, 8, 9}

my_set.discard(10)  # Discards an element if it exists; does nothing if it doesn't
print(my_set)       # Output: {1, 2, 4, 5, 6, 7, 8, 9}

popped_element = my_set.pop()  # Removes and returns an arbitrary element
print(popped_element)          # Output: (some element from the set)
print(my_set)                  # Output: Set without the popped element



Set Operations:
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}

# Union
print(set_a | set_b)  # Output: {1, 2, 3, 4, 5, 6}

# Intersection
print(set_a & set_b)  # Output: {3, 4}

# Difference
print(set_a - set_b)  # Output: {1, 2}

# Symmetric Difference
print(set_a ^ set_b)  # Output: {1, 2, 5, 6}


Membership Testing:
print(2 in set_a)  # Output: True
print(5 in set_a)  # Output: False


Iterating Through a Set:
for element in my_set:
    print(element)


Use Cases for Sets


Removing Duplicates from a List:
my_list = [1, 2, 2, 3, 4, 4, 5]
unique_elements = list(set(my_list))
print(unique_elements)  # Output: [1, 2, 3, 4, 5]


Membership Testing:
vowels = {'a', 'e', 'i', 'o', 'u'}
letter = 'e'
if letter in vowels:
    print(f"{letter} is a vowel")  # Output: e is a vowel


Set Operations for Data Analysis:
group_a = {'Alice', 'Bob', 'Charlie'}
group_b = {'Bob', 'David', 'Edward'}

# Students in either group
print(group_a | group_b)  # Output: {'Alice', 'Bob', 'Charlie', 'David', 'Edward'}

# Students in both groups
print(group_a & group_b)  # Output: {'Bob'}

# Students in group_a but not in group_b
print(group_a - group_b)  # Output: {'Alice', 'Charlie'}

# Students in group_a or group_b but not both
print(group_a ^ group_b)  # Output: {'Alice', 'Charlie', 'David', 'Edward'}


Summary
Unordered and Unique: Sets store unordered, unique elements.
Mutable: Sets can be modified by adding or removing elements.
Set Operations: Support mathematical operations like union, intersection, difference, and symmetric difference.
Use Cases: Ideal for removing duplicates, membership testing, and performing set operations in data analysis.
Sets are a powerful and flexible data structure in Python that can be particularly useful for tasks involving uniqueness and set operations.


'''
```

```python
# Q7. Describe how to add, modify and delete items in a dictionary with examples.

# Answer

'''


Dictionaries in Python
Dictionaries are a built-in data structure in Python that allow you to store data in key-value pairs. They are mutable, meaning you can add, modify, and delete items.

Adding Items
Adding a New Key-Value Pair:
my_dict = {"name": "Alice", "age": 25}
my_dict["city"] = "New York"
print(my_dict)  # Output: {"name": "Alice", "age": 25, "city": "New York"}


Using the update() Method:
my_dict.update({"email": "alice@example.com"})
print(my_dict)  # Output: {"name": "Alice", "age": 25, "city": "New York", "email": "alice@example.com"}


Modifying Items
Modifying an Existing Value:

my_dict["age"] = 26
print(my_dict)  # Output: {"name": "Alice", "age": 26, "city": "New York", "email": "alice@example.com"}

Using the update() Method:
my_dict.update({"name": "Alicia"})
print(my_dict)  # Output: {"name": "Alicia", "age": 26, "city": "New York", "email": "alice@example.com"}


Deleting Items
Using the del Statement:
del my_dict["city"]
print(my_dict)  # Output: {"name": "Alicia", "age": 26, "email": "alice@example.com"}


Using the pop() Method:
age = my_dict.pop("age")
print(age)      # Output: 26
print(my_dict)  # Output: {"name": "Alicia", "email": "alice@example.com"}

Using the popitem() Method:
Removes and returns the last inserted key-value pair.
last_item = my_dict.popitem()
print(last_item)  # Output: ("email", "alice@example.com")
print(my_dict)    # Output: {"name": "Alicia"}


Using the clear() Method:
Removes all items from the dictionary.

my_dict.clear()
print(my_dict)  # Output: {}


Summary
Adding Items: Use dict[key] = value or dict.update().
Modifying Items: Assign a new value to an existing key or use dict.update().
Deleting Items: Use del dict[key], dict.pop(key), dict.popitem(), or dict.clear().
These operations allow you to effectively manage the contents of a dictionary in Python.

'''
```

```python
# Q8. Discuss the importance of dictionary keys being immutable and provide examples.

# Answer

'''

In Python, dictionary keys must be immutable. This immutability requirement is crucial for maintaining the integrity and performance of dictionaries. 
Hereâ€™s why immutable keys are important, along with examples to illustrate these concepts:

Importance of Immutable Dictionary Keys


Hashability:
Dictionary keys must be hashable, meaning they must have a fixed hash value that remains the same during their lifetime.
Immutable types like strings, numbers, and tuples (containing only immutable elements) are hashable and can be used as dictionary keys.


Consistency:
If dictionary keys were mutable, their hash value could change, causing inconsistencies in key lookups and making it impossible to reliably retrieve stored values.
Example: If a list were used as a key, and the listâ€™s content changed, the dictionary would lose track of the original key.


Performance:
Hash tables, the underlying data structure for dictionaries, rely on quick lookups, insertions, and deletions based on consistent hash values.
Immutable keys ensure that hash values remain constant, allowing efficient data retrieval.


Examples of Immutable Keys

Using Strings as Keys:
my_dict = {"name": "Alice", "age": 25}
print(my_dict["name"])  # Output: Alice

Using Numbers as Keys:
my_dict = {1: "one", 2: "two", 3: "three"}
print(my_dict[2])  # Output: two

Using Tuples as Keys:
my_dict = {(1, 2): "a pair", (3, 4): "another pair"}
print(my_dict[(1, 2)])  # Output: a pair


Examples of Mutable Keys Causing Issues
Using a List as a Key (Raises TypeError):
my_dict = {}
key = [1, 2, 3]
# my_dict[key] = "a list"  # Raises TypeError: unhashable type: 'list'

Example of Mutable Keys Leading to Inconsistencies:
my_dict = {}
key = (1, 2, 3)
my_dict[key] = "a tuple"
key = (1, 2, 4)  # Changing the key would lead to loss of original key
# print(my_dict[key])  # Raises KeyError


Handling Complex Keys
If you need to use a mutable type as a key, you can convert it to an immutable type (like a tuple) before using it as a dictionary key.

Converting List to Tuple for Key:
my_dict = {}
key = [1, 2, 3]
immutable_key = tuple(key)
my_dict[immutable_key] = "a tuple converted from list"
print(my_dict)  # Output: {(1, 2, 3): 'a tuple converted from list'}

Using Frozenset for Sets as Keys:
my_dict = {}
key = frozenset([1, 2, 3])
my_dict[key] = "a frozenset"
print(my_dict)  # Output: {frozenset({1, 2, 3}): 'a frozenset'}


Summary
Hashability and Consistency: Immutable keys are necessary to ensure the hash values remain constant, allowing for consistent and efficient dictionary operations.
Performance: Immutable keys support the performance characteristics of hash tables, enabling quick lookups, insertions, and deletions.
Examples: Common immutable keys include strings, numbers, and tuples, while mutable types like lists cannot be used directly as dictionary keys.

'''
```
